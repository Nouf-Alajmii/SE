# 3. The A* Algorithm

<font size="4">The A* algorithm is an informed search type (use specific knowledge of a problem as a hint guide
it to the goal), the A* algorithm combines the characteristics of the uniform cost search (UCS) 
algorithm and the Greedy search algorithm to hit a target.
A* is complete and optimal. complete means that it guarantees to find a solution (if one exist), and 
optimal means that it guarantees to find the best path to a goal (lowest path cost).
A* search order by the sum: </font><br>
<font size="4">
<font color = "green">$$f(n)=g(n)+h(n)$$ </font>
</font>
<font size="4">
Where g(n) is a function that represents the cost of moving from the starting node to any another
node, and h(n) is a function that represents the heuristic estimated cost from node n to the goal 
node. We compute these functions and then add them to get the value of f(n), which to find the 
lowest cost in the neighboring node.</font><br>

3.2. تطبيق الخوارزمية على المشكلة
</font><br>
<font size="3">
    بعدما تعلمنا كيفية عمل خوارزمية A* سنطبقها على مشكلتنا
    <br></font></h2>

```python
class Node:
    def __init__(self, name, parent, h, cost): #make  constructor for class node
        self.name = name
        self.parent = parent
        self.h = h
        if parent :
            self.g = parent.g +cost
        else: 
            self.g = cost
        self.f = self.g + h
        
```

Node class has a constructor that creates node with following attributes:

state attribute: represents an object.

parent attribute: represents the previous object of state attribute.

h attribute: represent the heuristic of the state attribute to the goal(gun).

g attribute: represent the real cost path from the initial state to the state attribute, by finding the sum of the state attribute cost + the value of g in parent. Otherwise, it’s the initial state which mean it will be equal to 0.

f attribute: represent the evaluation function finding by the sum of g and h.


</font>
<h2 dir="rtl">
<font size="3">
تحتوي فئة العقدة: على مُنشئ يقوم بإنشاء عقدة بالسمات التالية

سمة الحالة: تمثل كائنًا

سمة الأصل: تمثل الكائن السابق لسمة الحالة

السمة h: تمثل الاستدلال على سمة الحالة للهدف (المسدس)

سمة g: تمثل مسار التكلفة الحقيقية من الحالة الأولية إلى سمة الحالة ، من خلال إيجاد مجموع تكلفة سمة الحالة + قيمة g في الأصل. بخلاف ذلك ، فإن الحالة الأولية هي التي تعني أنها ستكون مساوية للصفر

سمة f: تمثل دالة التقييم التي يتم إيجادها بمجموع g و h

```python
def add_node(name, parent): # -> make node for state
    node = Node(name,parent, heuristic1[name], graph_node[parent.name][name])
    return node
```

Method add_node has name and parent as a parameter, it is a method that creates a node by callen the Node class and passing the state, parent and the heuristic of the state from the dictionary, and the cost of the state from the dictionary.  Finally it return the created node to the caller.


</font>
<h2 dir="rtl">
<font size="3">
الطريقة add_node لها اسم وأصل كمعامل ، إنها طريقة تنشئ عقدة عن طريق استدعاء فئة العقدة وتمرير الحالة والأصل وإرشاد الحالة من القاموس ، وتكلفة الحالة من القاموس. أخيرًا ، تعيد العقدة التي تم إنشاؤها إلى المتصل.


```python
def expand(node): # -> expanding the node to access the neighbors nodes :
    neighbors = [] #create the neighbors list 
    for i in graph_node[node.name]:#for loop to appending node into the list
        neighbor = add_node(i, node)
        neighbors.append(neighbor)
    return neighbors #returing list
```


Method expand has one parameter (node), it is a method that expos all the neighbors
of the parameter by iterate. and add them to neighbors list and return the list.


</font>
<h2 dir="rtl">
<font size="3">
يحتوي أسلوب التوسيع على معلمة واحدة (عقدة) ، وهي طريقة تعرض جميع جيران المعلمة من خلال التكرار. وإضافتهم إلى قائمة الجيران وإرجاع القائمة.


```python
def A_star (start): #algorithm A* search
    
    # Create lists for open nodes and closed nodes
    expanded = []
    fringe = []
    # Add the start node
    fringe.append(start)
    
    # Loop until the open list is empty
    while len(fringe) > 0:
        # Sort the open list to get the node with the lowest cost first
        fringe.sort(key=lambda x: x.f)
        
        # Get the node with the lowest cost
        current_node = fringe.pop(0)
        
        # Add the current node to the closed list
        expanded.append(current_node)
        
        # Check if we found the goal , return the path
        if current_node.h == 0:
            path = [] #create path list
            
            # loop until we reach the beginning of the path
            while current_node != start:
                path.append(current_node.name)
                current_node = current_node.parent
                
            
            # Return reversed path
            return path[::-1] 
        # expanding the current node
        fringe.extend(expand(current_node)) 
      # if the fringe list length less or equal than 0  return "failure"
    return "failure"
```

Method A_star takes the start node as a parameter; it returns failure in case of no solution exist or the path when finding the solution; it inserts the parameter in the fringe and loop through the length of the fringe, if the fringe is empty it will return "failure". Otherwise, it will sort the fringe according to the lowest cost in front, then remove the front of the fringe, then it checks the node if it is the goal state it will return a path by backtracking of the current node until it reaches the start node. in case it is not the goal state it will insert all its neighbors to the fringe by invoking expand method. At the end of each iteration.
